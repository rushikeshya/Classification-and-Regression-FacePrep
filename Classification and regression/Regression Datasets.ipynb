# Regression Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

candydata = pd.read_csv("C:/Users/rushi/OneDrive/Desktop/Projects DS,ML,DL,AI/Candy bar prediction-Regression/candy-data.csv")
candydata
```

```python
x = candydata.iloc[:,[10,12]].values
y = candydata.iloc[:,9].values
```

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.20, random_state = 0)
```

```python
from sklearn.preprocessing import StandardScaler
sc_x =StandardScaler()
xtrain = sc_x.fit_transform(xtrain)
xtest = sc_x.transform(xtest)

print(xtrain[0:10])
```

# Logistic Regression Model

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state = 0)
classifier.fit(xtrain, ytrain)
```

```python
xtest
```

```python
y_pred = classifier.predict(xtest)
y_pred
```

```python
ytest
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(ytest, y_pred)

print("Confusion Matrix : \n", cm)
```

```python
from sklearn.metrics import accuracy_score
print("Accuracy : ",accuracy_score(ytest,y_pred))
```

# K-NN model

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors=10, metric='minkowski',p=2)
classifier.fit(xtrain,ytrain)
```

```python
y_pred = classifier.predict(xtest)
y_pred
```

```python
from sklearn.metrics import confusion_matrix  
cm= confusion_matrix(ytest, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
print("K-NN Accuracy : ",accuracy_score(ytest,y_pred))
```
